## Importing Packages

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
```

### Getting The DataFrames

```python
tips = sns.load_dataset('tips')
male_tips = tips[tips['sex']=='Male']
female_tips = tips[tips['sex']=='Female']
```

### Visualizing Data

```python
sns.relplot(x='total_bill', y='tip', data=female_tips)
sns.relplot(x='total_bill', y='tip', data=male_tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=female_tips)
sns.lmplot(x='total_bill', y='tip', data=male_tips)
```

```python
female_barplot=sns.barplot(x='total_bill', y='tip', data=female_tips)
female_barplot
```

```python
male_barplot =sns.barplot(x='total_bill', y='tip', data=male_tips)
male_barplot
```

```python
sns.pairplot(tips, hue="sex", markers=["o", "D"])
```

### Conclusions:

- Females May Consider Tiping More Than Males
- The Higher the Bill Customers May Consider Tiping More
- People Who Tip Do Exist XD
