## Importing Packages

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
```

### Getting The DataFrames

```python
df = pd.read_csv('./bestsellers_with_categories.csv')
df.head(5)
```

### Visualizing Data

```python
sns.relplot(y='User Rating', x='Price', data=df)
```

```python
sns.lmplot(y='User Rating', x='Price', data=df)
```

```python
sns.catplot(data=df, kind="violin", x="User Rating", y="Price", hue="Genre", split=True)
```
